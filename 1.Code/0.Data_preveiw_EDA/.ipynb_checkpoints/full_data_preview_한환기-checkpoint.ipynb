#### 카테고리 기준 데이터 특성 시각화
- impression_log.csv + viewer.csv + ad.csv + view_log, 노출광고별, 클릭여부, 클릭한 사람정보
- ad_log_viewer merge datasret 활용
- 카테고리(대분류/중분류)별 전

```python
ad_log_viewer = pd.read_csv('/Users/ppangppang/Desktop/ssac/Final_project/2.Processed_data/ad_log_viewer.csv')
ad_log_viewer
```

```python
# 카테고리 명칭 부여
cat_base = pd.read_excel('/Users/ppangppang/Desktop/ssac/Final_project/99.기타/광고 카테고리.xlsx')
cat_base = cat_base.rename(columns={'대분류' : 'category_id_1', '중분류' : 'category_id_2'})

ad_log_viewer = pd.merge(ad_log_viewer,cat_base,how='left',on=('category_id_1','category_id_2'))
ad_log_viewer
```

```python
ad_log = ad_log_viewer.sort_values(by='user_id_x')
ad_log.index = range(len(ad_log))
ad_log
```

```python
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
sns.countplot(data=ad_log_viewer, x='대분류명',hue='gender')
```

```python
# 연령 카테고리 지정
ad_log_viewer['age_range'] = ad_log_viewer['age']
for i in range(len(ad_log_viewer)):
    if ad_log_viewer['age_range'][i] <= 10:
        ad_log_viewer['age_range'][i] = '미식별자'
    elif ad_log_viewer['age_range'][i] <= 19:
        ad_log_viewer['age_range'][i] = '10대'
    elif ad_log_viewer['age_range'][i] <= 29:
        ad_log_viewer['age_range'][i] = '20대'
    elif ad_log_viewer['age_range'][i] <= 39:
        ad_log_viewer['age_range'][i] = '30대'
    elif ad_log_viewer['age_range'][i] <= 49:
        ad_log_viewer['age_range'][i] = '40대'
    elif ad_log_viewer['age_range'][i] <= 59:
        ad_log_viewer['age_range'][i] = '50대'
    elif ad_log_viewer['age_range'][i] <= 69:
        ad_log_viewer['age_range'][i] = '60대'        
    elif ad_log_viewer['age_range'][i] >= 70:
        ad_log_viewer['age_range'][i] = '미식별자'   
    else : 
        ad_log_viewer['age_range'][i] = '미식별자'   
```

```python
yclick = click[['age_range']].value_counts()
yclick = pd.DataFrame(data=yclick,columns=['age_y']).reset_index()
```

```python
nclick =ad_log_viewer[['age_range']].value_counts()
nclick = pd.DataFrame(data=nclick,columns=['age_n']).reset_index()
```

```python
rate_click = pd.merge(yclick,nclick,how='outer',on='age_range')
rate_click
```

```python
rate_click['age_r'] = round((rate_click['age_y'] /rate_click['age_n']),2)
rate_click
```

#### 컨텐츠별 등록 내용 누적 기록

```python
a=[]
for i in range(len(ad_log)-1):
    if ad_log['user_id_x'][i] != ad_log['user_id_x'][i+1]:
        a.append(i)
    else: pass
a
```

```python
ad_log['new']='0'
for i in a:
    if i==3:
        ad_log['new'][i] = ad_log['대분류명'][:i].cumsum()
    else:
        ad_log['new'][i] = ad_log['대분류명'][i-1:i].cumsum()
```
