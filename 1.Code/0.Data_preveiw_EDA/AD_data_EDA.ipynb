### 대분류

```python
# 대분류 키워드만 뽑아내는 함수 생성
def keyword_func(x):
    list1 = [] # 컬럼 1개에 들어 있는 내용들 담기
    list2 = [] # 컬럼마다 독립적으로 하니씩 끊어서 답기
    for i in range(len(x)):
        list1.append(x.keyword[i])
    
    for j in range(len(list1)):
        for i in range(len(str(list1[j]).split(','))):
            list2.append(str(list1[j]).split(',')[i]) 

    # 중복제거        
    keyword_list = [] 
    for i in range(len(list2)):
        if list2[i] not in keyword_list:
            keyword_list.append(list2[i])

    # 카테고리 번호, 리스트 반환
    return(('category_1_num : ',x.category_id_1.unique()),keyword_list)
```

```python
# 대분류 카테고리별 작업(카테고리만 뽑아오기)
ad_cat_list = ad.category_id_1.unique()
# 대분류 카테고리별 데이터 생성
ad_cat=[]
for i in range(len(ad_cat_list)):
    df = pd.DataFrame()
    df = ad[ad.category_id_1 == ad_cat_list[i]] 
    df.index = range(len(df)) # 인덱스 초기화
    ad_cat.append(df)
```

```python
# 대분류 카테고리별 고유 키워드 담기
ad_keyword_list=[]
for i in range(len(ad_cat)):
    ad_keyword_list.append(keyword_func(ad_cat[i]))
```

```python
# 대분류 카테고리별 head
test=pd.DataFrame()
for i in range(len(ad_cat)):
    test = pd.concat([test,ad_cat[i][:10]])
test
```
