# dataset의 이미지에 카테고리 붙이기

```python
dataset = pd.read_csv('/Users/ppangppang/Desktop/ssac/Final_project/0.Data_원본/data(ad)/dataset.csv')
dataset
```

```python
df = dataset.copy()
df = df.drop_duplicates('content_img_url')
df = df[['c_content_category_id_1','c_content_category_id_2','c_content_category_id_3','content_img_url']]
df = df.fillna(0)
df[['c_content_category_id_1','c_content_category_id_2','c_content_category_id_3']] =df[['c_content_category_id_1','c_content_category_id_2','c_content_category_id_3']].astype(int)
df.rename(columns={'c_content_category_id_1':'category_id_1','c_content_category_id_2':'category_id_2','c_content_category_id_3':'category_id_3'},inplace=True)
df.index = range(len(df))
df
```

```python
cat = pd.read_csv('/Users/ppangppang/Desktop/ssac/Final_project/2.Processed/data/ad_category_merge.csv')
cat = cat[['category_id_1','category_id_2','category_id_3','대분류명','중분류명']]
cat
```

```python
image_link = pd.merge(df,cat,how='left',on=('category_id_1','category_id_2','category_id_3'))
image_link = image_link[[ '대분류명', '중분류명','content_img_url']]
image_link = image_link.drop_duplicates('content_img_url')

# 대분류별 링크 정렬
image_link = image_link.sort_values(by='대분류명')
image_link.index=range(len(image_link))
image_link
```
