### 검색어로 나오는 판매글의 링크 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# 드라이버 실행
# driver = webdriver.Chrome()
```

```python
# 드라이버 실행
driver = webdriver.Chrome('/Users/ppangppang/Desktop/ssac/02.Python_/Python_Basic_Class/chromedriver')
```

```python
# cat 별 링크 들어가게 세팅
cat_list = ['아이폰', '갤럭시']
cat_name = ['iphone','galaxy']
pages = list(range(1, 3))
# 카테고리별 링크 담기
total_df = []

for cat in cat_list:
    # category별 dataframe 생성
    cat_df=pd.DataFrame()
    link=[]    
    for page in pages:
        url = f'https://m.bunjang.co.kr/search/products?q={cat}&order=score&page={page}'
        driver.get(url)
        time.sleep(2)
    
        # 페이지마다 사진 100장씩
        for i in range(1,101):
            try:
                link.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div/div[4]/div/div[{i}]/a').get_attribute('href'))
            except: pass
        print(f'{cat_name[cat_list.index(cat)]}항목 {page}페이지 완료')
        
    cat_df['link'] = link
    cat_df['category'] = cat_name[cat_list.index(cat)]
    total_df.append(cat_df)
    print(f'{cat_name[cat_list.index(cat)]}항목 작업 완료\n')
    
print('전체작업 완료')
```

```python
df_total = pd.DataFrame(columns = ['category', 'title', 'hashtag1', 'hashtag2', 'hashtag3', 'hashtag4', 'hashtag5', 'view', 'link'])

for i in range(2):
    title = []
    link = []
    view =[]
    hashtag1 = []
    hashtag2 = []
    hashtag3 = []
    hashtag4 = []
    hashtag5 = []
    category = []
    for j in range(200):
        url = total_df[i]['link'][j]
        driver.get(url)
        df = pd.DataFrame(columns = ['category', 'title', 'hashtag1', 'hashtag2', 'hashtag3', 'hashtag4', 'hashtag5', 'view', 'link'])

        try:
            # 글제목
            title.append(driver.find_element_by_xpath('//*[@id="root"]/div/div/div[4]/div[1]/div/div[2]/div/div[2]/div/div[1]/div[1]/div[1]').text)

            # 조회수
            view.append(driver.find_element_by_xpath('//*[@id="root"]/div/div/div[4]/div[1]/div/div[2]/div/div[2]/div/div[1]/div[2]/div[1]/div/div[2]').text)

            # 카테고리
            category.append(driver.find_element_by_xpath('//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[2]/div[2]/a/span').text)

            # 태그
            for k in range(1,6):
                if k == 1:
                    try:          
                        hashtag1.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[3]/div[2]/a[{k}]').text)
                    except:
                        hashtag1.append(0)
                elif k == 2:
                    try:
                        hashtag2.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[3]/div[2]/a[{k}]').text)
                    except:
                        hashtag2.append(0)
                elif k == 3:
                    try:
                        hashtag3.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[3]/div[2]/a[{k}]').text)
                    except:
                        hashtag3.append(0)
                elif k == 4:
                    try:
                        hashtag4.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[3]/div[2]/a[{k}]').text)
                    except:
                        hashtag4.append(0)
                elif k == 5:
                    try:
                        hashtag5.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[4]/div[1]/div[2]/div[1]/div[2]/div[3]/div[3]/div[2]/a[{k}]').text)
                    except:
                        hashtag5.append(0)

            # 사진링크
            link.append(driver.find_element_by_xpath(f'//*[@id="root"]/div/div/div[4]/div[1]/div/div[2]/div/div[1]/div/div[1]/img[1]').get_attribute('src'))
            df['category'] = category
            df['title'] = title
            df['hashtag1'] = hashtag1
            df['hashtag2'] = hashtag2
            df['hashtag3'] = hashtag3
            df['hashtag4'] = hashtag4
            df['hashtag5'] = hashtag5
            df['view'] = view
            df['link'] = link
        except: pass
    df_total = pd.concat([df_total, df])
```
