## 가중치 계산

```python
user = pd.read_csv('/Users/ppangppang/Desktop/ssac/Final_project/2.Processed_data/user_info.csv')
content= pd.read_csv('/Users/ppangppang/Desktop/ssac/Final_project/2.Processed_data/content_info.csv')
```

```python
content = content[['content_id', 'gender', 'age', '대분류명','노출정도']]
content['gender'] = round(content['gender'])
content['age'] = round(content['age'])
content
```

```python
# content별 클릭정도
click = click_con.groupby('content_id').sum().reset_index()
click
```

```python
click_con.head()
```

```python
content['content_id']=content['content_id'].astype(str)
user['user_id_x']=user['user_id_x'].astype(str)
```

```python
test = pd.merge(click_con,content,how='left',on='content_id')
test = pd.merge(test,user,how='left',on='user_id_x')
test = pd.merge(test,click,how='left',on='content_id')
test['click_rate'] = round(test['click_y']/test['노출정도'],2)
test = test[['content_id', 'user_id_x','click_x','click_rate','gender_x', 'age_x','대분류명','gender_y', 'age_y', 'cat1', 'cat2', 'cat3',
             'cat4','cat5']]
test.rename(columns={'click_x' : 'user_click', 'click_rate' : 'content_click_rate',
                     'gender_x':'content_gen_mean', 'age_x':'content_age_mean',
                     'gender_y':'user_gen','age_y':'user_age','대분류명':'content_cat'},inplace=True)
test
```

```python
# 성별,나이는 큰 점수부여 x, 카테고리, 유저 클릭에 가중치 크게 부여

# 성별 일치 : 10, 불일치 : 5
# 나이차이 * -2
# 대분류 1순위 일치 : 50, 2순위 일치 45, 3순위 일치 40.... 5순위까지만 해보기
# 상품당 클릭율 * 100
# 유저별 상품 클릭 횟수 * 50

test['gen_rate'] = 0.0
test['age_rate'] = 0.0
test['cat_rate'] = 0.0
test['con_click_rate'] = 0.0
test['user_click_rate'] =0.0

test['age_rate'] = abs(test['content_age_mean'] - test['user_age'])*(-2)
for i in range(len(test)):
    if test['content_gen_mean'][i] == test['user_gen'][i]:
        test['gen_rate'][i] = 10.0
    else:
        test['gen_rate'][i] = 5.0
    # 카테고리계산
    if test['cat1'][i] == test['content_cat'][i]:
        test['cat_rate'][i] = 50.0
    elif test['cat2'][i] == test['content_cat'][i]:
        test['cat_rate'][i] = 45.0
    elif test['cat3'][i] == test['content_cat'][i]:
        test['cat_rate'][i] = 40.0
    elif test['cat4'][i] == test['content_cat'][i]:
        test['cat_rate'][i] = 35.0
    elif test['cat5'][i] == test['content_cat'][i]:
        test['cat_rate'][i] = 30.0
```
