```python
def merge_df(df):
    # 데이터 merge
    df = total_df[0].copy()
    df.fillna(0,inplace=True)
    df.replace(0,'',inplace=True)
    df['merge'] = df['title'] + df['cat1'] + df['cat2'] + df['cat3']
    
    # 작업용 df로 변경
    df = df[['merge','keyword']]
    return df
```

### 명사만 추출

```python
info=[]
for i in range(len(adidas)):
    info.append(adidas['feature'][i])

from konlpy.tag import Okt
from collections import Counter

nouns_tagger = Okt()
nouns=[]
for line in info:
    nouns.append(nouns_tagger.nouns(line))
    
test = []
for i in range(len(nouns)):
    temp=''
    for j in range(len(nouns[i])):
        temp += nouns[i][j] + ' '
    test.append(temp)
info = test
info
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
countvect = vect.fit_transform(info)

countvect_df = pd.DataFrame(countvect.toarray(), columns = sorted(vect.vocabulary_))

from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(max_features = 10000)
tfvect = vect.fit(info)

tfidv_df = pd.DataFrame(tfvect.transform(info).toarray(), columns = sorted(vect.vocabulary_))
# tfidv_df.index = adidas['product_id']
tfidv_df
```
